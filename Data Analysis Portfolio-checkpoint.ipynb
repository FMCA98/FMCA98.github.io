```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
filename = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
```

```python
df.head()
```

```python
import numpy as np
```

```python
#want to get rid of ? 
df.replace('?',np.nan, inplace =True)
df.head()
```

```python
#find out how much and where the missing data is
missing_values = df.isnull()
```

```python
for i in df.columns:
    print (i)
    print (missing_values[i].value_counts())
    #this makes it more clear where values are missing 
```

```python
# next step is to deal with missing data, it can either be dropped or replaced
# Normalized losses, stroke, bore, horsepower and peak-rpm to be replaced by mean
# Num of doors is to be replaced by frequency-84% fo sedans have 4 doors
# drop values where price is missing as this is what we want to predict
avg_normalized_loss = df['normalized-losses'].astype('float').mean(axis=0)
df['normalized-losses'].replace(np.nan, avg_normalized_loss, inplace =True)
avg_bore = df['bore'].astype('float').mean(axis=0)
df['bore'].replace(np.nan, avg_bore, inplace=True) 
avg_horsepower = df['horsepower'].astype(float).mean(axis=0)
df['horsepower'].replace(np.nan, avg_horsepower, inplace = True)
avg_peak_rpm = df['peak-rpm'].astype(float).mean(axis=0)
df['peak-rpm'].replace(np.nan ,avg_peak_rpm, inplace = True)
avg_stroke = df['stroke'].astype(float).mean(axis= 0)
df['stroke'].replace(np.nan,avg_stroke, inplace = True) 
```

```python
#replace the missing values with the frequency
print(df['num-of-doors'].value_counts())
df['num-of-doors'].replace(np.nan, 'four', inplace = True)
```

```python
#check that values have been sucessfully changed 
df.dropna(subset=['price'], axis = 0 , inplace = True)
missing_values = df.isnull()
missing_values
```

```python
# the next step of the analysis process is to ensure the data is of the correct type
df[["bore", "stroke",]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses",'horsepower']] = df[["normalized-losses",'horsepower']].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
#normalize width, length and height 
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['length'] = df['length']/df['length'].max()
df['height'] = df['height']/df['height'].max()
```

```python
import matplotlib as plt
from matplotlib import pyplot
```

```python
plt.pyplot.hist(df['horsepower'])
plt.pyplot.xlabel('Horsepower')
plt.pyplot.ylabel('Value')
#horsepower is a continous variable and so group into low medium and high values via binning 
```

```python
#want to generate 3 bins of equal length 
avg_horsepower =df['horsepower'].astype('float').mean(axis=0)
df['horsepower'].replace(np.nan, avg_horsepower, inplace = True)
import numpy as np
HP = df['horsepower'].astype(int, copy=True)
bins_horsepower = np.linspace(min(HP), max(HP), 4)
```

```python
Group_names = ['Low','Medium','High']
HP_binned = pd.cut(HP, bins_horsepower, labels=Group_names, include_lowest=True )
```

```python

pyplot.bar(Group_names, HP_binned.value_counts())
plt.pyplot.xlabel('Horsepower')
plt.pyplot.ylabel('Value')
```

```python
#turn categorical values to numeric so that the feature can be used to predict price
Fuel_type = pd.get_dummies(df['fuel-type'])
Fuel_type.rename(columns = {'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
df  = pd.concat([df, Fuel_type], axis=1)

```

```python
df.drop('fuel-type', axis = 1, inplace =True)
df.head()
```

```python
#convert fuel type into numerical values
Aspiration=pd.get_dummies(df['aspiration'])
Aspiration.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace = True)
df= pd.concat([df,Aspiration], axis = 1)
```

```python
df.drop('aspiration', axis = 1, inplace = True)
df
```
